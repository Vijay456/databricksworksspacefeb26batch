```python
%sql
CREATE TABLE project1catalog.edw_gold.dim_category (
  CategoryId BIGINT GENERATED ALWAYS AS IDENTITY PRIMARY KEY,
  Category STRING,
  Subcategory STRING
)
USING delta
LOCATION 'abfss://datalake@adlsfeb26batch.dfs.core.windows.net/gold/dim_category/'
```

```python
%sql
CREATE TABLE project1catalog.edw_gold.dim_user (
  UserId BIGINT GENERATED ALWAYS AS IDENTITY PRIMARY KEY,
  OpenedBy STRING,
  SysCreatedBy STRING,
  SysUpdatedBy STRING
)
USING delta
LOCATION 'abfss://datalake@adlsfeb26batch.dfs.core.windows.net/gold/dim_user/'
```

```python
%sql
CREATE TABLE project1catalog.edw_gold.fact_event_log (
  Number STRING,
  UserId BIGINT,
  CallerId STRING,
  AssignmentGroup STRING,
  CategoryId BIGINT reference 
  Location STRING,
  CmdbCi STRING,
  Impact STRING,
  Urgency STRING,
  Priority STRING,
  IncidentState STRING,
  Active BOOLEAN,
  ReassignmentCount INT,
  ReopenCount INT,
  SysModCount INT,
  MadeSla BOOLEAN,
  OpenedAt TIMESTAMP,
  SysCreatedAt TIMESTAMP,
  SysUpdatedAt TIMESTAMP,
  ContactType STRING,
  USymptom STRING,
  AssignedTo STRING,
  Knowledge BOOLEAN,
  UPriorityConfirmation STRING,
  Notify STRING,
  ProblemId STRING,
  Rfc STRING,
  Vendor STRING,
  CausedBy STRING,
  ClosedCode STRING,
  ResolvedBy STRING,
  ResolvedAt TIMESTAMP,
  ClosedAt TIMESTAMP
)
USING delta
LOCATION 'abfss://datalake@adlsfeb26batch.dfs.core.windows.net/gold/fact_event_log/'
```
